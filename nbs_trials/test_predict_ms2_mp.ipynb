# Testing predict_mp()

As multiprocessing cannot run for classes defined in the same notebook (model_shop notebook), so we need to test multiprocessing functions using an external notebook.

```python
from peptdeep.model.rt import IRT_PEPTIDE_DF
from peptdeep.pretrained_models import ModelManager
import numpy as np

model_mgr = ModelManager(device='cuda')

df=IRT_PEPTIDE_DF.copy()
df['charge'] = 2
df['mods'] = ''
df['mod_sites'] = ''
idxes = np.zeros(len(df)+1,dtype=np.int64)
idxes[1:] = np.cumsum(df.nAA.values-1)
df['frag_start_idx'] = idxes[:-1]
df['frag_stop_idx'] = idxes[1:]
```

```python
ret_dict = model_mgr.predict_all(df, mp_batch_size=4,min_required_precursor_num_for_mp=4)
df = ret_dict['precursor_df']
df
```
