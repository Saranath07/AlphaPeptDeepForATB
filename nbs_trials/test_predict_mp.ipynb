# Testing predict_mp()

As multiprocessing cannot run for classes defined in the same notebook (model_shop notebook), so we need to test multiprocessing functions using an external notebook.

```python
from peptdeep.model.rt import IRT_PEPTIDE_DF
from peptdeep.model.model_shop import (
    Model_for_Generic_AASeq_Regression_LSTM,
    ModelInterface_for_Generic_AASeq_Regression
)
IRT_PEPTIDE_DF=IRT_PEPTIDE_DF.copy()
IRT_PEPTIDE_DF['detected_property'] = (
    IRT_PEPTIDE_DF.irt-IRT_PEPTIDE_DF.irt.min()
)/(IRT_PEPTIDE_DF.irt.max()-IRT_PEPTIDE_DF.irt.min())
model = ModelInterface_for_Generic_AASeq_Regression(
    model_class=Model_for_Generic_AASeq_Regression_LSTM
)
model.train(IRT_PEPTIDE_DF, epoch=20)

model.predict_mp(IRT_PEPTIDE_DF, mp_batch_size=3, process_num=2)
```
